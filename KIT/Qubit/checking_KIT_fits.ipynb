```python
L_F_eff_1 = 26.5
C_F_eff_1 = 26.9
EJ_1 = 5.4

L_F_eff_2 = 20.4
C_F_eff_2 = 22.4
EJ_2 = 9.5

L_F_eff_3 = 32.2
C_F_eff_3 = 24.8
EJ_3 = 5.6

φ_ext_values = np.linspace(0.5,1,100)

E_fluxonium_1_vs_φ_ext = np.array([sq_ext.diag(sq_ext.sq_fluxonium(EJ=EJ_1, L_F_eff=L_F_eff_1, C_F_eff=C_F_eff_1,φ_ext=φ_ext).hamiltonian(),n_eig=2)[0] for φ_ext in φ_ext_values])
E_fluxonium_2_vs_φ_ext = np.array([sq_ext.diag(sq_ext.sq_fluxonium(EJ=EJ_2, L_F_eff=L_F_eff_2, C_F_eff=C_F_eff_2,φ_ext=φ_ext).hamiltonian(),n_eig=2)[0] for φ_ext in φ_ext_values])
E_fluxonium_3_vs_φ_ext = np.array([sq_ext.diag(sq_ext.sq_fluxonium(EJ=EJ_3, L_F_eff=L_F_eff_3, C_F_eff=C_F_eff_3,φ_ext=φ_ext).hamiltonian(),n_eig=2)[0] for φ_ext in φ_ext_values])

ω01_fluxonium_1_vs_φ_ext = E_fluxonium_1_vs_φ_ext[:,1]-E_fluxonium_1_vs_φ_ext[:,0]
ω01_fluxonium_2_vs_φ_ext = E_fluxonium_2_vs_φ_ext[:,1]-E_fluxonium_2_vs_φ_ext[:,0]
ω01_fluxonium_3_vs_φ_ext = E_fluxonium_3_vs_φ_ext[:,1]-E_fluxonium_3_vs_φ_ext[:,0]
```

```python
fig, ax = plt.subplots(figsize=[4.5,3])
ax.plot(φ_ext_values, ω01_fluxonium_1_vs_φ_ext, 'b')
ax.plot(φ_ext_values, ω01_fluxonium_2_vs_φ_ext, 'g')
ax.plot(φ_ext_values, ω01_fluxonium_3_vs_φ_ext, 'r')

#ax.set_ylabel(r'$')
#ax.set_xlabel(r'$')
#ax.set_title(r'$')
ax.set_ylim([3,11])
ax.set_xlim([0.5,1])
fig.tight_layout()
fig.show()
```
