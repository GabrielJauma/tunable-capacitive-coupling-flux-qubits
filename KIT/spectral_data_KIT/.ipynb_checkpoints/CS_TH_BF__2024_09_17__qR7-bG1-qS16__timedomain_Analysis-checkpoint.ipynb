## Functions

```python
def line(x, c, y0):
    return c*x + y0

def exp_fit(t, A, tau, off):
    return A * np.exp(- t / tau) + off

def damped_oscillation(t, A, tau, f, phase, off):
    return A * np.exp(- t / tau) * np.cos(2 * np.pi * f * t + phase) + off

def damped_oscillation_beating(t, A, tau, f, df, phase, off):
    return A/2 * np.exp(- t / tau) * ( np.cos(2 * np.pi * (f-df/2) * t + phase) + np.cos(2 * np.pi * (f+df/2) * t + phase) )+ off

def damped_oscillation_gauss(t, A, tau, f, phase, off):
    return A * np.exp(- t**2 / tau**2) * np.cos(2 * np.pi * f * t + phase) + off


def iq_to_pop(i, q, ref_pos):
    dist_to_g = np.sqrt((i - ref_pos[0][0])**2 + (q - ref_pos[0][1])**2)
    dist_to_e = np.sqrt((i - ref_pos[1][0])**2 + (q - ref_pos[1][1])**2)
    
    return dist_to_g/(dist_to_g + dist_to_e)


def lorentzian( x, x0, a, gam, offset):
    return a * gam**2 / ( gam**2 + ( x - x0 )**2) + offset
```

## Write matrix

```python
# coil_vec = [coil1, coil2, coil3], qubit_vec = [qb1, qb2, qb3] = [qO15, qB1, qR7]

# inductance matrix: How many phi_0s per mA
L11 = +1/qR7__c1_I0
L12 = -1/qR7__c2_I0
L13 = -1/qR7__c3_I0

L21 = -1/bG1__c1_I0
L22 = +1/bG1__c2_I0
L23 = -1/bG1__c3_I0

L31 = -1/qS16__c1_I0
L32 = -1/qS16__c2_I0
L33 = +1/qS16__c3_I0

L = np.array([ [L11, +L12, -L13], [+L21, L22, -L23], [-L31, -L32, L33] ])

# offset vector: flux offset for each qubit 
phi_1_off = 1/2 - L11 * qR7__c1_Iss 
phi_2_off = 1/2 - L22 * bG1__c2_Iss 
phi_3_off = 1/2 - L33 * qS16__c3_Iss 

phi_off_vec = np.array([phi_1_off, phi_2_off, phi_3_off])
```

```python
Np = 20

qb1 = Flx.Fluxonium()
qb2 = Flx.Fluxonium()
qb3 = Flx.Fluxonium()

qb1.set_parameters(L=qR7__Lq_fit*1e-9, C=qR7__Cq_fit*1e-15, Ej=qR7__Ej_fit, N = Np)
qb1.calc_hamiltonian_parameters()

qb2.set_parameters(L=bG1__Lq_fit*1e-9, C=bG1__Cq_fit*1e-15, Ej = bG1__Ej_fit, N = Np)
qb2.calc_hamiltonian_parameters()

qb3.set_parameters(L=qS16__Lq_fit*1e-9, C=qS16__Cq_fit*1e-15, Ej = qS16__Ej_fit, N = Np)
qb3.calc_hamiltonian_parameters()

# phi and frequencies as look up tables for all coil operations: We calculate the frequencies only once 
# and not for each operation

phi_lookup = np.linspace(0.5, 1, 20001)

qb1.set_sweep_parameters(phi_lookup*2*np.pi)
qb1.sweep_external_flux()
qb2.set_sweep_parameters(phi_lookup*2*np.pi)
qb2.sweep_external_flux()
qb3.set_sweep_parameters(phi_lookup*2*np.pi)
qb3.sweep_external_flux()

E1 = qb1.E_sweep
f1_lookup = E1[:,:]-E1[0,:]
E2 = qb2.E_sweep
f2_lookup = E2[:,:]-E2[0,:]
E3 = qb3.E_sweep
f3_lookup = E3[:,:]-E3[0,:]
```

## Plot spectra

```python
def FKT__Ticks(Begin,End,Mod,Step,Fak = 1,Offset=0):
    
    if Fak == 1:
        Fak = 1
        while np.abs(Begin+Step)*Fak<1:
            Fak = Fak*10
        
    Begin = Begin * Fak
    End = End * Fak
    Mod = Mod * Fak
    Step = Step * Fak
    Offset = Offset * Fak
        
    List = []
    i=0
    while i <End-Begin+Step:
#        print(Begin+i+Offset,Mod,(Begin+i+Offset)%Mod)
        if (Begin+i+Offset)%Mod == 0: List.append((Begin+i)/Fak)
        else: List.append('')
        i = i+Step
        
#    print(List)

    for i in range (0,len(List)):
        if List[i]!='':
            if float(List[i])%1==0:
                List[i]=int(List[i])
#            List[i] = r'$'+str(List[i])+'$'
            List[i] = str(List[i])

    return(List)


TextSize=20

plt.figure(figsize=(10,6))

plt.plot(phi_lookup,f1_lookup[1].T,color='blue',label='qR7',lw=2)
plt.plot(phi_lookup,f2_lookup[1].T,color='green',label='bG1',lw=2)
plt.plot(phi_lookup,f3_lookup[1].T,color='red',label='qS16',lw=2)

plt.xlabel(r'$\Phi\ (\Phi_0)$',size=TextSize)
plt.xticks(np.arange(0,1.01,0.01),FKT__Ticks(0,1,0.05,0.01,Fak = 1,Offset=0),fontsize=TextSize)
plt.ylabel(r'$f_{\rm{q}}\ (\rm{GHz})$',size=TextSize)
plt.yticks(np.arange(0,11.1,0.1),FKT__Ticks(0,11,0.5,0.1,Fak = 1,Offset=0),fontsize=TextSize)

plt.legend(fontsize=TextSize)

plt.xlim([0.5,1.0])
plt.ylim([3,11])

plt.grid()
#plt.savefig('calibrated_Qubits__qO15_bG1_qR7.jpg')
plt.show()
```
